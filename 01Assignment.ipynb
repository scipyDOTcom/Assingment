### Q3) Three Coins are tossed, find the probability that two heads and one tail are obtained?


```python
outcome=["H","T"]

sample_space=[]

for i in outcome:
    for j in outcome:
        for k in outcome:
            sample_space.append(i+j+k)

sample = []
for i in sample_space:
    if i.count("H")==2:
        sample.append(i)

print("Total Probability:-",sample_space,len(sample_space))
print("Two Head and One Tail probability :- ",sample,len(sample))

```

### Q4 Two Dice are rolled, find the probability that sum is 

### a)	Equal_to_1
### b)	Less than or equal to 4
### c)	Sum is divisible by 2 and 3


```python

dice_int1 = 1,2,3,4,5,6

dice_int2 = []
for i in dice_int1:
    for j in dice_int1:
        dice_int2.append(i+j)
        

print(dice_int2,len(dice_int2))
equal_to_zero =[]
for i in dice_int2:
    if i ==0:
        equal_to_zero.append(i)
print("equal_to_zero:-",equal_to_zero,len(equal_to_zero))

lessthan4 = []

for i in dice_int2:
    if i<=4:
        lessthan4.append(i)
print("Less than or equal to 4:- ",lessthan4,len(lessthan4))


divisible2_3 = []

for i in dice_int2:
    if i%(2*3)==0:
        divisible2_3.append(i)
print("divisible by 2 and 3 :-",divisible2_3,len(divisible2_3))
    
```
